#### This project seeks to explore, segment, and cluster the neighborhoods in the city of Toronto. However, the neighborhood data is not readily available on the internet, a Wikipedia page exists that has all the information needed exist. Data will be scraped from the Wikipedia page, wrangled and cleaned and then read into a pandas dataframe so that it is in a structured format.

## Part One 

### The Data scraping process

```python
# importing the libraries
from bs4 import BeautifulSoup
import requests
import csv

#sending a HTTP request to a url
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
# Make a GET request to fetch the raw HTML content
page = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(page, "lxml")
# print(soup.prettify())  # print the parsed data of html
# get the table having the class wikitable
match = soup.find("table", class_="wikitable sortable")

#writing a CSV file
headers = ["Postal_code", "Borough", "Neighborhood"]
filename = "Canada_Postal_Code.csv"
with open(filename, "w") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(headers)

# Analysing the HTML tag
    for row in match.select('tbody tr'):
        row_text = [x.text for x in row.find_all('td')]
    #print(row_text)
        for i in row_text:
            postal_code = row_text[0].strip()
            print (postal_code)
        
            borough =row_text[1].strip()
            print(borough)
                
            neighborhood=row_text[2].strip()
            print(neighborhood)
    
            print()
    
        csv_writer.writerow([postal_code, borough, neighborhood]) 
csvfile.close()
    
    
```
